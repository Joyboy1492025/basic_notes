# Python Variables

## 1. Introduction to Python
Python is a high-level, interpreted language.

## 2. Variable in Python
- A **variable** is a name that refers to a value.
- It acts as a container for data.

## 3. Characteristics of Variables
- **Dynamic Typing**: No need to declare types.
- **No Explicit Declaration**: Assign to create.
- **Mutable**: Values can change during execution.

## 4. Rules to Define Variables
- Must start with a letter (a–z, A–Z) or underscore.
- Can contain letters, digits (0–9), and underscores.
- Case‑sensitive (`age`, `Age`, and `AGE` are different).
- Cannot be a Python keyword.

### Syntax
```python
variable_name = value


````python
age = 25
print(age)  # Output: 25

name = "Alice"
print(name)  # Output: Alice
````

````python
first_number = 242
second_number = 92
sum_of_2_no = first_number + second_number
print(sum_of_2_no)
````

````python
# Multiple Assignments 

intvar , floatvar , strvar = 10,2.57,"Python Language" # Using commas to separat 
print(intvar) 
print(floatvar) 
print(strvar) 
````

````python
# All variables pointing to same value
p1 = p2 = p3 = p4 = 44  
print(p1,p2,p3,p4) 
````

# Python Data Types

- ### Numeric
  - int, float, complex

- ### Sequence
   - list, tuple, range

- ### Text
   - str

- ### Mapping
   - dict

- ### Set Types
   - set, frozenset

- ### Boolean
   - bool
     
**note:** `type()` function use for detecting a data type of Variables

````python
# -------------------------------
# Python Data Types Demonstration
# -------------------------------

# Numeric types
int_data = 24
float_data = 583.25
complex_data = 4 + 9j

# Text type
str_data = "This is a String"

# Sequence types
list_data = [1, 2, 'hi']
tuple_data = (1, 2, 3)   # tuple uses parentheses

# Mapping type
list_dict = {
    'Name': 'OM',
    'Age': 24,
    'City': 'Goa'
}

# Set types
set_data = {1, 2, 3, 2}          # duplicates are removed
frozenset_data = frozenset([1, 2, 3, 2])

# Boolean type
bool_data_true = True
bool_data_false = False

# -------------------------------
# Print type information
# -------------------------------
print(f'This is a {type(int_data)}')        # <class 'int'>
print(f'This is a {type(float_data)}')      # <class 'float'>
print(f'This is a {type(complex_data)}')    # <class 'complex'>
print(f'This is a {type(str_data)}')        # <class 'str'>
print(f'This is a {type(list_data)}')       # <class 'list'>
print(f'This is a {type(tuple_data)}')      # <class 'tuple'>
print(f'This is a {type(list_dict)}')       # <class 'dict'>
print(f'This is a {type(set_data)}')        # <class 'set'>
print(f'This is a {type(frozenset_data)}')  # <class 'frozenset'>
print(f'This is a {type(bool_data_true)}')  # <class 'bool'>
````

````python
# -------------------------------
# Print values to verify
# -------------------------------
print("Integer value:", int_data)
print("Float value:", float_data)
print("Complex value:", complex_data)
print("String value:", str_data)
print("List value:", list_data)
print("Tuple value:", tuple_data)
print("Dictionary value:", list_dict)
print("Set value:", set_data)
print("Frozenset value:", frozenset_data)
print("Boolean True value:", bool_data_true)
print("Boolean False value:", bool_data_false)
````

# 🔑 Python Data Types Classification

## 1. Primitive vs Non‑Primitive

- **Primitive Data Types**  
  Basic building blocks of Python. or single value
  Examples:  
  - `int` → integers  
  - `float` → decimal numbers  
  - `complex` → complex numbers  
  - `bool` → True/False  
  - `str` → strings  

- **Non‑Primitive Data Types**  
  More advanced structures that can hold multiple values.  
  Examples:  
  - `list`  
  - `tuple`  
  - `dict`  
  - `set`  
  - `frozenset`

---

## 2. Hashable (Hexable) vs Unhashable (Unhexable)

- **Hashable (Hexable)**  
  - Can be used as dictionary keys and hash value remains constant  
  - Must be **immutable**.  
  - Examples: `int`, `float`, `str`, `tuple`, `frozenset`, `bool`.

- **Unhashable (Unhexable)**  
  - Cannot be used as dictionary keys.  
  - Usually **mutable**.  
  - Examples: `list`, `dict`, `set`.

---

## 3. Immutable vs Mutable

- **Immutable Data Types**  
  - Values cannot change after creation.  
  - Examples:  
    - `int`, `float`, `complex`, `str`, `tuple`, `frozenset`, `bool`

- **Mutable Data Types**  
  - Values can change after creation.  
  - Examples:  
    - `list`, `dict`, `set`
