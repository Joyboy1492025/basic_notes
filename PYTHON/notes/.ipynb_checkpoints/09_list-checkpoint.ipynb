## 📊 Structures in Python

| Structure   | Dimension | Example in Python | Explanation |
|-------------|-----------|-------------------|-------------|
| **Scalar**  | 0D        | `x = 5` | A single value (integer, float, string). No axes. |
| **Vector**  | 1D        | `pd.Series([1,2,3])` | A one-dimensional array. Represents a list of values. |
| **Matrix**  | 2D        | `np.array([[1,2],[3,4]])` | A two-dimensional grid (rows × columns). |
| **DataFrame** | 2D (with labels) | `pd.DataFrame({"A":[1,2], "B":[3,4]})` | Like a labeled matrix with row/column names. Useful for tabular data. |
| **Tensor**  | nD        | `np.array([[[1,2],[3,4]], [[5,6],[7,8]]])` | A generalized n-dimensional array (can be 3D, 4D, etc.). Used in deep learning. |

---

## 🔹 Scalar (0D)
- **Definition:** Just a single number or value.
- **Python Example:**
  ```python
  x = 42        # integer scalar
  y = 3.14      # float scalar
  ```
- **Use Case:** Represents constants like age, temperature, or a single measurement.

---

## 🔹 Vector (1D)
- **Definition:** A sequence of numbers (like a list).
- **Python Example:**
  ```python
  import numpy as np
  v = np.array([1, 2, 3])   # vector
  ```
- **Use Case:** Represents features of a dataset, directions, or simple collections.

---

## 🔹 Matrix (2D)
- **Definition:** A rectangular grid of numbers with rows and columns.
- **Python Example:**
  ```python
  M = np.array([[1, 2, 3],
                [4, 5, 6]])
  ```
- **Use Case:** Linear algebra, image pixels (grayscale), tabular data.

---

## 🔹 DataFrame (2D with labels)
- **Definition:** A labeled 2D table from **pandas**.
- **Python Example:**
  ```python
  import pandas as pd
  df = pd.DataFrame({
      "Name": ["Alice", "Bob"],
      "Age": [25, 30]
  })
  ```
- **Use Case:** Data analysis, machine learning datasets, CSV/Excel-like operations.

---

## 🔹 Tensor (nD)
- **Definition:** A generalization of scalars, vectors, and matrices to higher dimensions.
- **Python Example:**
  ```python
  T = np.array([[[1, 2], [3, 4]],
                [[5, 6], [7, 8]]])   # 3D tensor
  ```
- **Use Case:** Deep learning (images = 3D tensors: width × height × channels; videos = 4D tensors: frames × width × height × channels).

---

## 🧠 Dimensional Intuition
- **Scalar (0D):** A single point.  
- **Vector (1D):** A line of points.  
- **Matrix (2D):** A flat grid (like a spreadsheet).  
- **Tensor (nD):** A cube or hypercube of data.  

---

✨ In short:  
- Scalars are single values.  
- Vectors are lists.  
- Matrices are tables.  
- DataFrames are labeled tables.  
- Tensors are multi-dimensional generalizations, crucial in AI/ML.  

## 🔹 1. Data Types
These are the **basic building blocks** — single values.

- **Numeric types** → `int`, `float`, `complex`
- **Boolean type** → `bool`
- **Text type** → `str`

👉 Example:
```python

y = 3.14      # float
z = 2 + 3j    # complex
flag = True   # bool

```

---

## 🔹 2. Data Structures
These are **collections of data types** — they hold more than one value.

### ✅ Built-in Data Structures
- **List** → Ordered, mutable collection  
  ```python
  my_list = [1, 2, 3, "hello"]
  ```
- **Tuple** → Ordered, immutable collection  
  ```python
  my_tuple = (1, 2, 3)
  ```
- **Set** → Unordered, unique elements  
  ```python
  my_set = {1, 2, 3}
  ```
- **Dictionary** → Key-value pairs  
  ```python
  my_dict = {"name": "Alice", "age": 25}
  ```

### ✅ User-Defined Data Structures
You can also **create your own structures** using classes:
```python
class Student:
    def __init__(self, name, age):
        self.name = name
        self.age = age

s1 = Student("Alice", 20)
```
Here, `Student` is a **user-defined data structure** that groups multiple values (name, age).

## 📋 Common List Methods

| Method | Description | Example |
|--------|-------------|---------|
| `append(x)` | Add an item to the end | `lst.append(5)` |
| `extend(iterable)` | Add all items from another iterable | `lst.extend([6,7])` |
| `insert(i, x)` | Insert item at position `i` | `lst.insert(1, "hi")` |
| `remove(x)` | Remove first occurrence of value | `lst.remove(3)` |
| `pop([i])` | Remove and return item at index (last by default) | `lst.pop()` |
| `clear()` | Remove all items | `lst.clear()` |
| `index(x)` | Return index of first occurrence | `lst.index(20)` |
| `count(x)` | Count occurrences of value | `lst.count(2)` |
| `sort(key=None, reverse=False)` | Sort list in place | `lst.sort()` |
| `reverse()` | Reverse list in place | `lst.reverse()` |
| `copy()` | Return a shallow copy | `lst2 = lst.copy()` |

---

## 🧠 Key Notes
- **In-place vs. copy:** Methods like `sort()`, `reverse()`, `clear()` modify the list directly. Methods like `copy()` return a new list.
- **Errors:** `remove(x)` and `index(x)` raise `ValueError` if the item isn’t found. `pop(i)` raises `IndexError` if index is out of range.
- **Flexibility:** Lists can hold mixed data types, unlike arrays in other languages.

## list comprehension
[expression for item in iterable if condition]


````python
list3 = [i for i in range(1,11)]
print(list3)
````

````python
print([i*i for i in range(1,11)])
````

````python
print([i for i in range(1,11) if i % 2 ==0])
````

````python
words = ["apple", "banana", "cherry"]
caps = [w.upper() for w in words]
print(caps)   # ['APPLE', 'BANANA', 'CHERRY']
````

````python
pairs = [(x, y) for x in [1, 2] for y in [3, 4]]
print(pairs)   # [(1, 3), (1, 4), (2, 3), (2, 4)]
````

````python
num = [1,2,3,4,5]
print([nums*2 for nums in num])
````

````python
print(5 in num)
````

````python
print(9 in num)
````

````python
num.count()
````

````python
num = [int(x) for x in input("Enter a numbers :").split()]
print(num)
print(type(num))
````

## All / Any 
 
The `all()` method returns: 
 
**True** - If all elements in a list are true. 
**False** - If any element in a list is false.
 
The `any()` function returns True if any element in the list is True. If not, any() returns False.

````python
all([True,True,True])
````

````python
all([True,True,True,False])
````

````python
any([True,True,True,False])
````

````python
any([1,2,3,True] )
````

````python
s = 'ada12'
any(s2.isalnum() for s2 in s)
````

````python
s = [1,2,3,4,5,6,7,8]
all(s2 % 2 ==0 for s2 in s)
````
