**Dictionary (`dict`)** is a **mutable, unordered collection** of key–value pairs.  

### 🔹 Key Features
- **Mutable** → You can add, remove, or update items after creation.
- **Keys must be immutable** → Examples: strings, numbers, tuples.  
- **Values can be anything** → Any Python object (mutable or immutable).
- **Unordered** → Items don’t have a fixed position (though since Python 3.7, insertion order is preserved for iteration).

---

### 🔹 Common Dictionary Methods
| Method        | Description |
|---------------|-------------|
| `.keys()`     | Returns all keys |
| `.values()`   | Returns all values |
| `.items()`    | Returns key-value pairs |
| `.get(key)`   | Safely get value (returns `None` if key not found) |
| `.update()`   | Update dictionary with another dict |
| `.pop(key)`   | Remove and return value of given key |
| `.popitem(key)`   | Remove and return item |
| `.clear()`    | Remove all items |

---

👉 So, to summarize:
- **dict is mutable** (you can change it anytime).
- **Keys must be immutable types**.
- **Values can be any type**.

## Dictionary Methods

```python
dit1 = {'name': 'Alice', 'age': 21, 'course': 'Python', 'grade': 'A'}
```

```python
dit1.keys() #Returns all keys
```

```python
dit1.values() #Returns all values
```

```python
dit1.items() #Returns key-value pairs
```

```python
dit1.get('name') #Safely get value (returns None if key not found)
```

```python
dit1.pop('grade')
```

```python
dit1.update({'grade':'A'})
dit1
```

```python
dit1.popitem() # Removes and returns the last inserted key–value pair
dit1
```

### Iterating over dictionary


```python
# Iterating over dictionary
for key, value in dit1.items():
    print(key, ":", value)
```

```python
for i in dit1: 
    print(i)
```

```python
for i in dit1: 
    print(i , ':' , dit1[i])
```

```python
21 in dit1
```

```python
dit1.clear()
```

```python
del dit1
```

### Dict comprehension
[key: value  for item in iterable if condition]


```python
print( {x:x for x in range(1,11)} )
```

```python
print( {f'Item {x}':x for x in range(1,11)} )
```
