### 🔹 `range()`
- Generates a sequence of numbers.
- Commonly used in loops.
- Syntax: `range(start, stop, step)`

```python
print( [ i for i in range(9) ] )
```

```python
print( [ i for i in range(1,10) ] )
```

```python
print( [ i for i in range(1,50,5) ] )
```

### 🔹 `random` module
- Used for generating random numbers.
- You need to `import random` first.
- Common functions:
  - `random.randint(a, b)` → random integer between `a` and `b` (inclusive).
  - `random.random()` → random float between `0.0` and `1.0`.
  - `random.choice(seq)` → random element from a sequence.
  - `random.shuffle(seq)` → shuffle a list in place.

```python
import random
```

```python
print( random.randint(1,10) )
```

```python
print( random.random() )
```

```python
l1 = [x for x in range(1,6)]
random.shuffle(l1) 
l1
```

```python
l2 = [x for x in range(1,6)]
print( random.choice(l2) ) 
l2
```


### 🔹 Commonly Used Functions in `random`
| Function | Description | Example |
|----------|-------------|---------|
| `random.random()` | Returns a float between `0.0` and `1.0`. | `random.random()` → `0.3745` |
| `random.randint(a, b)` | Returns a random integer between `a` and `b` (inclusive). | `random.randint(1, 10)` → `7` |
| `random.randrange(start, stop, step)` | Returns a random number from `range(start, stop, step)`. | `random.randrange(1, 10, 2)` → `3` |
| `random.uniform(a, b)` | Returns a random float between `a` and `b`. | `random.uniform(1, 5)` → `3.42` |
| `random.choice(seq)` | Returns a random element from a sequence. | `random.choice(['apple','banana','cherry'])` → `'banana'` |
| `random.choices(seq, k=n)` | Returns a list of `n` random elements (with replacement). | `random.choices([1,2,3], k=2)` → `[2, 3]` |
| `random.sample(seq, k=n)` | Returns `n` unique random elements (without replacement). | `random.sample([1,2,3,4], 2)` → `[1, 4]` |
| `random.shuffle(seq)` | Shuffles a list in place. | `random.shuffle(my_list)` |
| `random.seed(n)` | Sets the seed for reproducibility. | `random.seed(42)` |


```python
import random

# Random float
print("random():", random.random())

# Random integer
print("randint(1, 10):", random.randint(1, 10))

# Random range
print("randrange(1, 10, 2):", random.randrange(1, 10, 2))

# Random uniform float
print("uniform(1, 5):", random.uniform(1, 5))

# Random choice
fruits = ["apple", "banana", "cherry"]
print("choice:", random.choice(fruits))

# Random multiple choices
print("choices:", random.choices(fruits, k=2))

# Random sample
print("sample:", random.sample(fruits, 2))

# Shuffle
random.shuffle(fruits)
print("shuffled:", fruits)
```
