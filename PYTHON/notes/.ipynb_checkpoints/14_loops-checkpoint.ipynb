## Loops in Python  

A **loop** lets you repeat a block of code multiple times, either for a fixed number of iterations or until a condition is met. Python mainly uses **`for` loops** and **`while` loops**.

| Loop Type | Structure | Example | Output |
| --- | --- | --- | --- |
| **For Loop** | ``for var in sequence:`` | ``for i in range(5): print(i)`` | 0,1,2,3,4 |
| **While Loop** | ``while condition:`` | ``num = 1 while num <= 5: print(num) num += 1`` | 1,2,3,4,5 |
| **Nested Loop** | ``for ...:\\n for ...:`` | ``for i in range(2): for j in range(3): print(i, j)`` | (0,0),(0,1),(0,2),(1,0),(1,1),(1,2) |

---

### **`For` Loop**
Used to iterate over a sequence (like a list, tuple, string, or range).
```python
for i in range(5):
    print(i)
```

---


````python
for i in range(1,6): 
    print(i)
````

````python
for i in range(6,0,-1): #Reverse Looping
    print(i)
````

````python
for i in range(1,11):
    print(f"{ i } = { i*i }")
````

````python
for i in "HELLO":
    print(i)
````

````python
li1 = [43,23,53,123,532] #for list,tuple,set

for i in li1:
    print(i)
````

````python
for i in enumerate(range(11,16)):
    print(i)

print("-----------------")

for i,j in enumerate(range(11,16)):
    print(i,j)
````

````python
dict1 = {"name":"om",
        "age":34,
        "city":"goa",
        "last_name":"joy" 
        }

for i in dict1:
    print(i)

print("-----------------")

for i in dict1.items():
    print(i)    

print("-----------------")
    

for i,j in dict1.items():
    print(f"Key :{i}, Value :{j}")       
````

### **`While` Loop**
Repeats as long as a condition is `True`.
```python
num = 1
while num <= 5:
    print(num)
    num += 1
```
---



````python
num = 1
while num <= 5:
    print(num)
    num += 1
````

````python
#while True:
#    print("It wil print unlimited")
````

````python
num = int(input("Enter number of itretion :"))
num1 = 1
while num1 <= num:
    print(num1)
    num1 += 1
````

````python
while True:
    if num == 0:
        break
    num = int(input("Enter number :"))
    print("Yes don't want to contine than enter 0")
    print("Even" if num % 2 == 0 else "Odd")
````

````python
# Reverse Looping
num = 10
while num >= 0:
    print(num)
    num -= 1
````

## 3. **Nested Loops**
Loops inside loops — useful for working with matrices or repeated patterns.
```python
for i in range(3):
    for j in range(2):
        print(f"i={i}, j={j}")
```
---



````python
for i in range(3): # it run three time 
    for j in range(2): # it run two time
        print(f"i={i}, j={j}")
````

````python
for i in range(5): # it run three time 
    for j in range(3): # it run two time
        print(i,j)
````

### 4. **Loop Control Statements**
- **`break`** → exits the loop immediately.
- **`continue`** → skips the current iteration and moves to the next.
- **`else` with loops** → runs if the loop finishes normally (without `break`).
- **`pass`**  → Do nothing (placeholder)
  
```python
for i in range(5):
    if i == 3:
        break
    print(i)
else:
    print("Loop finished without break")
```
---

    

````python
# break exits the loop immediately
for i in range(1,11):
    if i == 4:
        break # end when 4 appear
    print(i)
````

````python
# continue skips the current iteration and moves to the next.
for i in range(1,11):
    if i == 4: 
        continue  # 4 skip and continue
    print(i)
````

````python
# pass Do nothing (placeholder)
for i in range(1,11):
    if i == 4: 
        pass # use to prevent error
    print(i)
````

````python
# else with loops → runs if the loop finishes normally (without break).
for i in range(5):
    if i == 3:
        break
    print(i)
else:
    print("Loop finished without break")
````

````python
for i in range(5):
    if i == 9:
        break
    print(i)
else:
    print("Loop finished without break")
````

---
## ⚡ Key Takeaways
- Use **`for`** when iterating over a sequence.
- Use **`while`** when repeating until a condition changes.
- Combine with **`break`** and **`continue`** for fine control.
- **Nested loops** are powerful but can get complex quickly.
