# 📝 Python Strings

## 1. Introduction
- A **string** is a sequence of characters enclosed in single quotes `'...'` or double quotes `"..."`.
- Strings are **immutable** → once created, they cannot be changed.

## 2. Creating Strings
- **Single quotes**: 'Python'
- **Double quotes**: "Python"
- **Triple quotes**: '''Python''' or """Python""" (for multi-line strings)

```python
text = 'My'

print(text + 'World')      # Concatenation
print(text * 10)           # Repetition
print(type(text))          # Type check
print(len(text))           # Length of string
```

**note**: `len()` Function don't work with number data

## String Slicing 

```python
str1 = "PYTHON"

print('str1[:]   =', str1[:])       # whole string
print('str1[1:4] =', str1[1:4])     # from index 1 up to (but not including) 4
print('str1[4:]  =', str1[4:])      # from index 4 to end
print('str1[:4]  =', str1[:4])      # from start up to (but not including) 4
print('str1[:100]=', str1[:100])    # safe slicing beyond length
```

```python
str1 = "PYTHON"

print('str1[-4:]   =', str1[-4:])     # from index -4 to end
print('str1[-4:-1] =', str1[-4:-1])   # from index -4 up to (but not including) -1
```

```python
str1 = "PYTHON"

print('str1[0:7:2] =', str1[0:7:2])   # slice with step = 2
print('str1[::-1]  =', str1[::-1])    # reverse string
print('str1[::2]   =', str1[::2])     # every 2nd character
print('str1[2::]   =', str1[2::])     # from index 2 to end
```

## 🔹 Common String Methods in Python

| Method | Description | Example | Output |
|--------|-------------|---------|--------|
| `upper()` | Converts to uppercase | `"hello".upper()` | `"HELLO"` |
| `lower()` | Converts to lowercase | `"HELLO".lower()` | `"hello"` |
| `title()` | Capitalizes each word | `"python string methods".title()` | `"Python String Methods"` |
| `capitalize()` | Capitalizes first letter | `"python".capitalize()` | `"Python"` |
| `strip()` | Removes whitespace (or chars) | `"  hello  ".strip()` | `"hello"` |
| `lstrip()` | Removes leading whitespace | `"   hello".lstrip()` | `"hello"` |
| `rstrip()` | Removes trailing whitespace | `"hello   ".rstrip()` | `"hello"` |
| `replace(a,b)` | Replace substring | `"apple".replace("a","A")` | `"Apple"` |
| `split()` | Splits into list | `"a,b,c".split(",")` | `['a','b','c']` |
| `join()` | Joins list into string | `",".join(['a','b','c'])` | `"a,b,c"` |
| `find()` | Finds index of substring | `"hello".find("e")` | `1` |
| `index()` | Like `find()` but error if not found | `"hello".index("l")` | `2` |
| `startswith()` | Checks prefix | `"hello".startswith("he")` | `True` |
| `endswith()` | Checks suffix | `"hello".endswith("lo")` | `True` |
| `count()` | Counts occurrences | `"banana".count("a")` | `3` |
| `isdigit()` | Checks if all chars are digits | `"123".isdigit()` | `True` |
| `isalpha()` | Checks if all chars are letters | `"abc".isalpha()` | `True` |
| `isalnum()` | Checks if all chars are letters/digits | `"abc123".isalnum()` | `True` |
| `isspace()` | Checks if all chars are spaces | `"   ".isspace()` | `True` |
| `swapcase()` | Swaps case | `"Hello".swapcase()` | `"hELLO"` |
| `zfill(width)` | Pads with zeros | `"42".zfill(5)` | `"00042"` |
| `center(width)` | Centers string | `"hi".center(6)` | `"  hi  "` |
| `ljust(width)` | Left-aligns string | `"hi".ljust(6)` | `"hi    "` |
| `rjust(width)` | Right-aligns string | `"hi".rjust(6)` | `"    hi"` |

```python
s = "  Python String Methods  "

print(s.upper())        # PYTHON STRING METHODS
print(s.lower())        # python string methods
print(s.strip())        # Python String Methods
print(s.replace("Python", "Java"))  #   Java String Methods  
print(s.split())        # ['Python', 'String', 'Methods']
print("-".join(["A","B","C"]))      # A-B-C
print(s.find("String")) # 9
print(s.startswith("  Py")) # True
print(s.count("t"))     # 3
```
