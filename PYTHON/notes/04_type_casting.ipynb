# Type casting
Type casting in Python is the process of converting a variable from one data type to another. This is especially useful when you need to perform operations between different types or ensure compatibility in your code.
- **Implicit Type Conversion** (Type Casting done automatically)
    - eg: int + float → result becomes float
- **Explicit Type Conversion** (Type Casting done manually)
    - eg: int(num_str)

### other to int

```python
# Explicit type casting examples

# Float → Integer (decimal part truncated, not rounded)
print(int(23.4))   # Output: 23

# Boolean → Integer (True becomes 1, False becomes 0)
print(int(True))   # Output: 1

# String → Integer (string must contain valid digits)
print(int('23'))   # Output: 23
```

```python
print(int('23.4'))   # ❌ ValueError: invalid literal for int() with base 10: '23.4'
```

```python
print(int(2+6j))   # ❌ TypeError: can't convert complex to int
```

```python
print(int(2.5, 6.3))  
# ❌ TypeError: int() can't convert non-string with explicit base
```

### other to Float

```python
# Explicit type casting to float

# Integer → Float
print(float(43))       # Output: 43.0

# Boolean → Float (True → 1.0, False → 0.0)
print(float(False))    # Output: 0.0

# Float → Float (no change)
print(float(23.43))    # Output: 23.43

# String (decimal) → Float
print(float('23.43'))  # Output: 23.43

# String (integer) → Float
print(float('23'))     # Output: 23.0
```

```python
print(float(52+5j)) # ❌ TypeError: can't convert complex to float
```

```python
print(float(52,5)) # ❌ TypeError: float() can't convert non-string with explicit base
```

### other to Complex


```python
# Explicit type casting to complex numbers

# Integer → Complex (imag part defaults to 0)
print(complex(5))            # Output: (5+0j)

# Float → Complex (imag part defaults to 0)
print(complex(5.54))         # Output: (5.54+0j)

# String (integer) → Complex
print(complex('14'))         # Output: (14+0j)

# String (float) → Complex
print(complex('14.54'))      # Output: (14.54+0j)

# Two numeric arguments → Complex (real, imag)
print(complex(25, 36))       # Output: (25+36j)

# Two float arguments → Complex (real, imag)
print(complex(2.2, 5.6))     # Output: (2.2+5.6j)

# Boolean → Complex (True → 1+0j, False → 0+0j)
print(complex(True))         # Output: (1+0j)

# Boolean pair → Complex (real=True → 1, imag=False → 0)
print(complex(True, False))  # Output: (1+0j)
```

```python
print(complex('2.3', '5.6'))
# ❌ TypeError: complex() can't take second arg if first is a string
```

```python
print(complex(2.2, 5.6, 2.6))
# ❌ TypeError: complex() takes at most 2 arguments (3 given)
```

```python
print(complex('2.3', '5.6'))
# ❌ TypeError: complex() can't take second arg if first is a string
```

```python
print(complex('True'))
# ❌ ValueError: complex() arg is a malformed string
```

### other to bool

```python
# Explicit type casting to boolean
# Rule: In Python, bool() returns False only for 0, 0.0, 0j, empty collections ([], {}, ()), empty strings "", and None.
# Everything else evaluates to True.

print(bool(1))        # True → non-zero integer
print(bool(1.4))      # True → non-zero float
print(bool(9+2j))     # True → non-zero complex number
print(bool('hi'))     # True → non-empty string
print(bool('25.62'))  # True → non-empty numeric string
```

```python
# Explicit type casting to boolean
# Rule: bool() returns False only for 0, 0.0, 0j, empty collections ([], {}, ()), empty strings "", and None.
# Everything else evaluates to True.

print(bool())       # False → no argument defaults to 0
print(bool(0))      # False → zero integer
print(bool(None))   # False → None is treated as "null" → False
print(bool(""))     # False → empty string
print(bool([]))     # False → empty list
```

```python
print(bool(1,2))
# ❌ TypeError: bool() takes at most 1 argument (2 given)
```

 ### other to string

```python
# Explicit type casting to string
# Rule: str() converts any Python object into its string representation.

print(str(1))        # "1" → integer to string
print(str(1.2))      # "1.2" → float to string
print(str(3+4j))     # "(3+4j)" → complex number to string
print(str(True))     # "True" → boolean to string
print(str([1,2,3]))  # "[1, 2, 3]" → list to string
```

```python
print(str(1,2))
# ❌ TypeError: str() takes at most 1 argument (2 given)
```

### Others 

```python
# Converting between built-in collections

# Tuple → List
print(list((1, 2, 3, 4)))
# Output: [1, 2, 3, 4]
# Explanation: list() takes an iterable (tuple here) and creates a list.

# Tuple → Set (duplicates removed automatically)
print(set((1, 2, 3, 4, 4, 5, 2)))
# Output: {1, 2, 3, 4, 5}
# Explanation: set() takes an iterable and creates a set, eliminating duplicates.

# Set → Tuple
print(tuple({1, 2, 3, 4, 3, 2}))
# Output: (1, 2, 3, 4)
# Explanation: tuple() takes an iterable (set here) and creates a tuple.
# Note: Sets are unordered, so the order of elements in the tuple may vary.
```

```python
# From list of tuples (key-value pairs)
print(dict([('a', 1), ('b', 2)]))
# Output: {'a': 1, 'b': 2}
```
