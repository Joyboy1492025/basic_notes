# Input and Output functions

| Function | Purpose | Example | Output |
|----------|---------|---------|--------|
| `print()` | Displays output | `print("Hi")` | `Hi` |
| `input()` | Takes user input | `x = input("Enter: ")` | User types something → stored in `x` |

---



## print() Function
The `print()` function is used to display output on the screen.


```python
print("Hello World")          # Prints a string
print(10)                     # Prints an integer
print(3.14)                   # Prints a float
print("Sum =", 5+3)           # Prints text and result together
print("A", "B", "C", sep="-") # Custom separator → A-B-C
print("Line1\nLine2")         # Newline inside string
print("End", end="***")       # Custom end → End***
```

### Key parameters
- sep → defines separator between multiple values (default is space).
- end → defines what to print at the end (default is newline \n).

### Print formating

```python
print(f"I am aditya and i am {24} years old") # format eauation in it
```

### Format Specifiers in F-Strings

| Specifier | Meaning | Example | Output |
|-----------|---------|---------|--------|
| `:.2f` | 2 decimal places | `f"{3.14159:.2f}"` | `3.14` |
| `:.0f` | No decimals | `f"{3.14159:.0f}"` | `3` |
| `:d` | Integer | `f"{42:d}"` | `42` |
| `:x` | Hexadecimal | `f"{255:x}"` | `ff` |
| `:b` | Binary | `f"{8:b}"` | `1000` |
| `:>10` | Right align width 10 | `f"{'hi':>10}"` | `"        hi"` |
| `:<10` | Left align width 10 | `f"{'hi':<10}"` | `"hi        "` |
| `:^10` | Center align width 10 | `f"{'hi':^10}"` | `"    hi    "` |


```python
print(f"I am aditya and i am {24.2425242:.2f} years old") # format eauation float
print(f"I am aditya and i am {24.2425242:.3f} years old") # format eauation float
print(f"I am aditya and i am {24.2425242:.0f} years old") # format eauation float
```

```python
print(f"Binary : {25:b}")
print(f"Hexadecimal : {25:x}")
print(f"Octal : {25:o}")
```

## input() Function
The `input()` function is used to take input from the user. It always returns a **string**.


```python
num = input("Enter a number=")# Since input() returns a string, you often need type casting
print( num + num )
print(type(num))
```

### type casting for input

```python
a = int(input("Enter a number="))
print( a*a )
```

```python
a = float(input("Enter a number="))
print( a*a )
```

```python
a = complex(input("Enter a number="))
print( a*a )
```

```python
a = round(input("Enter a number="))
print( a*a )
```

### multi input

```python
a,b = input("Enter name and surname: ").split()
print(f"you first name is {a} and surname is {b}")
```

```python
# in int()
a,b = map(int , input("Enter name and surname: ").split()) 
print(f"Sum of a and b is {a+b}")
```
