## 🔹 Parameter Tuning
- **By default, the system learns parameters automatically.**
- These are the **internal values** (like weights in a neural network, regression coefficients, or decision tree splits) that the algorithm adjusts during training.
- You don’t set them manually — the training process (gradient descent, optimization) does the work.
- **Analogy:** In Windows, the system itself decides internal values to run smoothly — you don’t manually tweak them.

---

## 🔹 Hyperparameter Tuning
- **Hyperparameters are external settings you (the user/developer) configure before training.**
- Examples: learning rate, number of epochs, batch size, number of neighbors in KNN, max_depth in decision trees.
- These control *how* the model learns, not the values it learns.
- **Analogy:** If Windows 10 didn’t fit your needs and you switched to Windows 11, that’s like changing a hyperparameter — you’re altering the system’s configuration, not the internal learned values.

---

## 🌟 Key Difference
- **Parameter tuning** → automatic, inside the model (system decides).  
- **Hyperparameter tuning** → manual, outside the model (you decide).  


# Tuple
A **tuple** in Python is another built-in data structure, very similar to a list, but with one key difference: **tuples are immutable**. That means once you create a tuple, you cannot change, add, or remove its elements.

---

## 🔹 Characteristics of Tuples
- **Ordered** → Elements maintain their position.
- **Immutable** → Cannot be modified after creation.
- **Heterogeneous** → Can store mixed data types (like lists).
- **Indexed** → Supports indexing and slicing.

---

## 📊 Comparison: List vs Tuple

| Feature        | List (`[]`) | Tuple (`()`) |
|----------------|-------------|--------------|
| Mutability     | Mutable (can change) | Immutable (cannot change) |
| Syntax         | `[1, 2, 3]` | `(1, 2, 3)` |
| Performance    | Slower (more flexible) | Faster (fixed structure) |
| Use Case       | Dynamic data | Fixed data (constants, keys) |

---

## 🌟 Why Use Tuples?
- To store **fixed collections** of items (like coordinates `(x, y)`).
- As **dictionary keys** (lists cannot be used as keys because they’re mutable).
- For **performance optimization** when data doesn’t need to change.

# set
### 🔑 Key Features of Sets
- **No duplicates**: Each element appears only once.
- **Unordered**: Items don’t have a fixed position or index.
- **Mutable**: You can add or remove elements.
- **Efficient operations**: Great for membership tests and mathematical set operations.

👉 However, the **elements inside a set must be immutable** (like numbers, strings, or tuples). You cannot put a list or another set inside a set because those are mutable.
So, to summarize:
- **Set itself → mutable**
- **Elements inside → must be immutable**

---
### ⚙️ Common Methods
| Method | Description | Example |
| --- | --- | --- |
| ``add()`` | Adds an element | ``my_set.add(5)`` |
| ``remove()`` | Removes an element (error if not found) | ``my_set.remove(2)`` |
| ``discard()`` | Removes an element (no error if not found) | ``my_set.discard(10)`` |
| ``pop()`` | Removes a random element | ``my_set.pop()`` |
| ``clear()`` | Empties the set | ``my_set.clear()`` |
| ``union()`` | Combines sets | ``a.union(b)`` |
| ``update()`` | Adds elements from another set/iterable | ``a.update(b)`` |
| ``intersection()`` | Common elements | ``a.intersection(b)`` |
| ``intersection_update()`` | Keeps only common elements (in place) | ``a.intersection_update(b)`` |
| ``difference()`` | Elements in one but not the other | ``a.difference(b)`` |
| ``difference_update()`` | Removes elements found in another set (in place) | ``a.difference_update(b)`` |
| ``symmetric_difference()`` | Elements in either but not both | ``a.symmetric_difference(b)`` |
| ``symmetric_difference_update()`` | Updates set with elements in either but not both | ``a.symmetric_difference_update(b)`` |
| ``issubset()`` | Checks if all elements are in another set | ``a.issubset(b)`` |
| ``issuperset()`` | Checks if set contains all elements of another | ``a.issuperset(b)`` |
| ``isdisjoint()`` | Checks if sets have no common elements | ``a.isdisjoint(b)`` |
---
### 🚀 When to Use Sets
- Removing duplicates from a list.
- Fast membership checks (`if x in my_set`).
- Performing mathematical set operations (union, intersection, etc.).
- Symbols:
   -   **union() => |**
   -   **intersection() => &**
   -   **difference() => -**
   -   **symmetric_difference() => ^**
---

### 🔹 `frozenset` in Python
- **Immutable version of a set** → Once created, you cannot add, remove, or change elements.
- **Hashable** → Because it’s immutable, a `frozenset` can be used as a key in a dictionary or as an element inside another set (unlike normal sets).
- **Unique elements only** → Just like a regular set, duplicates are automatically removed.
- **Unordered** → Elements don’t have a fixed position.

### 🔹 Difference Between `set` and `frozenset`

| Feature            | `set` (mutable) | `frozenset` (immutable) |
|--------------------|-----------------|--------------------------|
| Can add/remove     | ✅ Yes          | ❌ No                   |
| Hashable           | ❌ No           | ✅ Yes                  |
| Use as dict key    | ❌ No           | ✅ Yes                  |
| Elements unique    | ✅ Yes          | ✅ Yes                  |


```python
fs = {1,2,32,42,24,53,23,3}
fs
```

```python
fs1 = frozenset(fs)
type(fs1)
```

### Generator
- Similar to list comprehension but produces a generator (lazy evaluation).
- It doesn’t store all results at once—each value is generated only when needed.

```python
# Even squares only
even_squares = (x*x for x in range(1, 11))
print(next(even_squares))  
print(next(even_squares))  
print(next(even_squares))  
```
