## Conditional Statements
Conditional statements in Python allow programs to make decisions by executing code only when certain conditions are met. 

---

### 🧩 Types of Conditional Statements
| Statement Type | Example | Use Case |
|----------------|---------|----------|
| **If** | `if x < 0: print("Negative")` | Runs only when condition is true |
| **If-Else** | `if age <= 12: print("Child") else: print("Adult")` | Two possible outcomes |
| **If-Elif-Else** | `if age <= 12: ... elif age <= 19: ... else: ...` | Multiple conditions checked sequentially |
| **Nested If** | `if age >= 60: if is_member: print("30% discount")` | Condition inside another condition |
| **Ternary (shorthand)** | `status = "Adult" if age >= 18 else "Minor"` | Compact one-line conditional |
| **Match-Case** | ```python match number: case 1: print("One") case _: print("Other")``` | Similar to switch-case in other languages  [GeeksForGeeks](https://www.geeksforgeeks.org/python/conditional-statements-in-python/) |

---

In Python, **indentation** is absolutely critical because it defines the structure of your code blocks. Unlike many other languages that use braces `{}` or keywords to mark blocks, Python uses spaces (or tabs, but spaces are strongly recommended).  

---

### 🔑 Why Indentation Matters
- Every block of code under `if`, `for`, `while`, `def`, `class`, etc. must be indented.
- The standard is **4 spaces per indentation level**.
- Misaligned indentation will raise an `IndentationError`.

---

### ⚠️ Common Mistakes
- Mixing tabs and spaces → leads to errors.
- Forgetting to indent after `if`, `for`, `while`, etc.
- Over‑indenting unnecessarily.

---

### ✅ Best Practices
- Always use **spaces, not tabs** (PEP 8 recommends 4 spaces).
- Configure your editor/IDE to insert spaces automatically.
- Keep indentation consistent across your project.
- `minumum 1 space required`

## `if - elif - else` Statements
It allows you to check multiple conditions in sequence, and only one branch of code will execute.

## 🔑 Structure
```python
if condition1:
    # code runs if condition1 is True
elif condition2:
    # code runs if condition2 is True (only if condition1 was False)
elif condition3:
    # code runs if condition3 is True (only if previous were False)
else:
    # code runs if none of the above conditions are True
```
---

````python
str1 = input("Enter any thing :")

if str1.isalpha():
    print("You enter alphabet data")
elif str1.isnumeric():
    print("You enter numeric data")
elif str1.isalnum():
    print("You enter alphanumeric data")
else:
    print("This data is out of if")
````

````python
age1 = int(input("Enter a age :"))

if age1 < 13:
    print("Child")
elif age1 < 18:
    print("Teenager")
elif age1 < 60:
    print("Adult")
else:
    print("Senior Citizen")
````

## Ternary Statements
The ternary operator in Python is a compact way to write conditional expressions. Instead of using a full if-elif-else block, you can assign values or return results in a single line.

```python
value_if_true if condition else value_if_false
```
---

````python
num4 = 4
result = "Even" if num4 % 2 == 0 else "Odd"
print(result)
````

````python
num5 = int(input("Enter a Number :"))
print( "positive number" if num5 > 0 else "negitive number" )
````

````python
score = int(input("Enter a Number :"))
grade = "A" if score >= 90 else "B" if score >= 75 else "C"
print(grade)
````

## Nested If
A **nested `if`** means putting one `if` statement inside another. It’s useful when you need to check **a condition only if another condition is already true**.  

## 🔑 Structure
```python
if condition1:
    # code block for condition1
    if condition2:
        # code block for condition2 (only runs if condition1 is True)
    else:
        # runs if condition2 is False, but condition1 was True
else:
    # runs if condition1 is False
```

---


````python
age = 65
is_member = True

if age >= 60:
    print("Senior citizen")
    if is_member:
        print("Eligible for 30% discount")
    else:
        print("Eligible for 20% discount")
else:
    print("Not a senior citizen")
````

````python
num = input("Enter a Number :")

if num.isnumeric():
    print("You enter a numeric value")
    if int(num) % 2 == 0:
        print(f"{num} is Even")
    else:
        print(f"{num} is Odd")
else:
    print("You enter a non-numeric value")
    
````

## `match-case` in Python (introduced in Python 3.10)

The **`match-case`** statement is Python’s version of a **switch-case** found in other languages. It’s used for pattern matching, making code cleaner when handling multiple possible values or structures.

## 🔑 Structure
```python
match variable:
    case pattern1:
        # code runs if variable matches pattern1
    case pattern2:
        # code runs if variable matches pattern2
    case pattern3:
        # code runs if variable matches pattern3
    case _:
        # default case (matches anything not handled above)
```

---


````python
day = "Monday"

match day:
    case "Monday":
        print("Start of the week")
    case "Friday":
        print("End of the work week")
    case "Sunday":
        print("Weekend rest")
    case _:
        print("Regular day")
````

````python
point = (0, 5)

match point:
    case (0, 0):
        print("Origin")
    case (0, y):
        print(f"On Y-axis at {y}")
    case (x, 0):
        print(f"On X-axis at {x}")
    case (x, y):
        print(f"Point at ({x}, {y})")
````

````python
option = int(input("Enter a option (1,2,3,4)"))

match option:
    case 1 | 'a':
        print("Python")
    case 2 | 'a':
        print("R")       
    case 3 | 'a':
        print("C/C++")
    case 4 | 'a':
        print("Web Dev")
    case _ :
        print("I don't want to became programmer")
````
