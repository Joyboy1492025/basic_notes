### 🔑 Key Points about Functions
- Defined using the `def` keyword.
- Can take **parameters** (inputs).
- Can return **values** using `return`.
- Can be called multiple times with different arguments.

### 📌 Types of Functions
1. **Built-in functions** — Already available in Python (e.g., `len()`, `print()`, `sum()`).
2. **User-defined functions** — Functions you create with `def`.
3. **Lambda functions** — Small anonymous functions using `lambda`.

### ⚡ Example: Arithmetic Function
```python
def add_numbers(a, b):
    return a + b

result = add_numbers(5, 7)
print("Sum:", result)
```
**Output:**
```
Sum: 12
```

### ✅ Why Use Functions?
- **Reusability**: Write once, use many times.
- **Modularity**: Break complex problems into smaller parts.
- **Readability**: Code is easier to understand.
- **Maintainability**: Easier to update and debug.

### Function with parameter / argument

| Aspect | Formal Arguments | Actual Arguments |
| --- | --- | --- |
| **Definition** | Variables in function definition | Values passed during function call |
| **Existence** | Inside the function | Outside the function |
| **Role** | Placeholder for input | Real input values |
| **Example** | ``name, ``age`` in ``def ``greet(name, ``age)`` | ``"Joy", ``25`` in ``greet("Joy", ``25)`` |

````python
# Function with parameter / argument
def add(x,y): # createion of function name add with 2 argument
    print(x+y) # x,y is formal argument

add(2,8) # call that function
add(75,23) # 75,23 actual argument
````

````python
# Function with parameter / argument
def add(x,y):
    c = x+y
    print(c) 

r = add(2,8) 
print(r)
````

````python
def add(x,y):
    c = x+y
    return c 

r = add(2,8) 
print(r)
````

### print vs return
| Feature | ``print`` | ``return`` |
| --- | --- | --- |
| Shows output on screen | ✅ Yes | ❌ No |
| Passes value back to program | ❌ No | ✅ Yes |
| Useful for | Debugging, user-facing messages | Computation, reuse of results |
| Can be assigned to variable | ❌ No | ✅ Yes |

````python
def greet(): # createion of function name greet
    return 'hello good morning'

def add(x,y):
    c = x+y
    return c 

r1 = greet()
r2 = add(2,6)
print(r1)
print(r2)
print(type(r1))
print(type(r2))
````

````python
def add_sub(x,y):
    a = x + y
    b = x- y
    return a,b
c = add_sub(8,3)   
print(c)    # if in return we pas more that 2 than it is in tuple
print(type(c))
````

## Other argument (Actual Arguments)

| Type | Symbol | Example Call | Stored As |
| --- | --- | --- | --- |
| Positional | ``func(a,b)``| ``func(10, 20)`` | Ordered |
| Keyword | ``func(a,b)`` | ``func(a=10, b=20)`` | Named |
| Default | ``func(a,b=20)`` | ``func(10)`` (uses default for b) | Predefined |
| Variable (*args) | ``func(*a)`` | ``func(1,2,3,4)`` | Tuple |
| Variable (**kwargs) | ``func(**a)`` | ``func(x=1, y=2)`` | Dictionary |

````python
def users(name,age,sex,edu='BE',skill = 'Python'):
    return [name,age,sex,edu,skill]
````

````python
# This is a postional arguments (work with position)
print(users('om',24,'male','BCOM','SQL'))
````

````python
# This is a keyword arguments (work with keyword)
print(users(age=29,sex='Female',name='Liza',skill='CMD',edu='MTech'))
````

````python
# This is a default arguments (work with keyword)
print(users('om',24,'male')) # edu and skill is default
````

## Variable (*args) And (**kwargs) arguments

#### 🔹 `*args` (Non-keyword variable arguments)
- Collects **extra positional arguments** into a **tuple**.
- Useful when you don’t know how many values will be passed.

#### 🔹 `**kwargs` (Keyword variable arguments)
- Collects **extra keyword arguments** into a **dictionary**.
- Useful when you want flexible named inputs.


````python
def demo(a, b=10, *args, **kwargs):
    print("a:", a)
    print("b:", b)
    print("args:", args)
    print("kwargs:", kwargs)

demo(1, 2, 3, 4, x=5, y=6)
````

````python
def total(*m):
    x = 0
    for i in m:
        x = x + i
    return x

print(total(90,34,56))    
````

````python
def account(name, age, *skill, **social):
    print(f'Name is {name}')
    print(f'Age is {age}')
    for i in skill:
        print(f'skill is {i}')
    for j,k in social.items():    
        print(f'social is {j} {k}')

account('Aditya',24, 'python' ,'sql','c',phone= 123456789,whatsapp= 12348956894)    
````

## Lambda function

#### 🔹 What is a Lambda Function?
- A **lambda function** is an **anonymous function** (no name) defined using the `lambda` keyword.
- Syntax:
  ```python
  lambda arguments: expression
  ```
- It can take any number of arguments but only **one expression** (which is automatically returned).

---

#### 🔹 Key Points
- Lambdas are **single-expression functions**.
- They return values automatically (no `return` keyword needed).
- Best used for **short, simple operations**.
- For complex logic, stick with `def`.

---


````python
# 1. Basic Lambda
squ = lambda x : x * x
print(squ(5))
````

````python
# 2. Multiple Arguments
add = lambda x,y : x + y
print(add(2,5))
````

````python
# 3. With `map()`
nums = [1, 2, 3, 4]
squares = list(map(lambda x: x**2, nums))
print(squares)   # Output: [1, 4, 9, 16]

nums1 = ['1','3','4','6']
t1 = list(map(int , nums1))
print(t1,type(t1[0]))
````

````python
# 4. With `filter()`

nums = [10, 15, 20, 25, 30]
evens = list(filter(lambda x: x % 2 == 0, nums))
print(evens)   # Output: [10, 20, 30]
````

````python
# 5. With `reduce()`
from functools import reduce

nums = [1, 2, 3, 4]
product = reduce(lambda x, y: x * y, nums)
print(product)
````
