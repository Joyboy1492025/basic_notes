# Complete Jupyter Notebook Markdown Guide 🚀

A full beginner-to-advanced guide for using Markdown inside Jupyter Notebook.

---

# 1. Headings

```markdown
# H1 Heading
## H2 Heading
### H3 Heading
#### H4 Heading
##### H5 Heading
###### H6 Heading
````

# H1 Heading

## H2 Heading

### H3 Heading

#### H4 Heading

##### H5 Heading

###### H6 Heading

---

# 2. Text Formatting

## Bold

```markdown
**Bold Text**
__Bold Text__
```

**Bold Text**

## Italic

```markdown
*Italic Text*
_Italic Text_
```

*Italic Text*

## Bold + Italic

```markdown
***Bold + Italic***
```

***Bold + Italic***

## Strikethrough

```markdown
~~Deleted Text~~
```

~~Deleted Text~~

## Inline Code

```markdown
`print("Hello")`
```

`print("Hello")`

## Highlight

```html
<mark>Highlighted Text</mark>
```

<mark>Highlighted Text</mark>

---

# 3. Lists

## Bullet List

```markdown
- Item 1
- Item 2
  - Nested Item
    - Deep Nested Item
```

* Item 1
* Item 2

  * Nested Item

    * Deep Nested Item

## Numbered List

```markdown
1. First
2. Second
3. Third
```

1. First
2. Second
3. Third

## Task List

```markdown
- [x] Completed
- [ ] Pending
```

* [x] Completed
* [ ] Pending

---

# 4. Links

## External Link

```markdown
[Google](https://google.com)
```

[Google](https://google.com)

## Internal Link

```markdown
[Jump to Tables](#7-tables)
```

[Jump to Tables](#7-tables)

---

# 5. Horizontal Line

```markdown
---
```

---

# 6. Blockquotes

```markdown
> Main quote
>> Nested quote
```

> Main quote
>
> > Nested quote

---

# 7. Tables

```markdown
| Name | Age | City |
|---|---|---|
| John | 25 | London |
| Sara | 22 | Paris |
```

| Name | Age | City   |
| ---- | --- | ------ |
| John | 25  | London |
| Sara | 22  | Paris  |

## Alignment Table

```markdown
| Left | Center | Right |
| :--- | :---: | ---: |
| A | B | C |
```

| Left | Center | Right |
| :--- | :----: | ----: |
| A    |    B   |     C |

---

# 8. Mathematics (LaTeX)

## Inline Formula

```markdown
$E = mc^2$
```

$E = mc^2$

## Block Formula

```markdown
$$
a^2 + b^2 = c^2
$$
```

$$
a^2 + b^2 = c^2
$$

## Matrix

```markdown
$$
\begin{bmatrix}
1 & 2 \\
3 & 4
\end{bmatrix}
$$
```

$$
\begin{bmatrix}
1 & 2 \
3 & 4
\end{bmatrix}
$$

---

# 9. Code Blocks

## Python

````markdown
```python
def hello():
    print("Hello")
```
````
