```r
roy <- readline("Enter Vote for roy :")
sam <- readline("Enter Vote for roy :")
joy <- readline("Enter Vote for roy :")

roy <- as.integer(roy)
sam <- as.integer(sam)
joy <- as.integer(joy)

total = roy + sam + joy
print(paste("Total Votes is",total))
```

```r
roy <- as.integer( readline("Enter Vote for roy :") )
sam <- as.integer( readline("Enter Vote for roy :") )
joy <- as.integer( readline("Enter Vote for roy :") )

total = sum(roy, sam, joy)
print(paste("Total Votes is",total))
```
